```python
def linear_kernel(X):
    return np.dot(X, X.T)

def polynomial_kernel(X, gamma=1, coef0=1, degree=3):
    return (gamma * np.dot(X, X.T) + coef0) ** degree

def rbf_kernel(X, gamma=1):
    X = X.values
    sq_dists = np.sum(X**2, axis=1).reshape(-1, 1) + np.sum(X**2, axis=1) - 2 * np.dot(X, X.T)
    return np.exp(-gamma * sq_dists)

linear_k = linear_kernel(X_train)
poly_k = polynomial_kernel(X_train, gamma=0.1, coef0=1, degree=3)
rbf_k = rbf_kernel(X_train, gamma=0.1)

print("Linear Kernel Shape:", linear_k.shape)
print("Polynomial Kernel Shape:", poly_k.shape)
print("RBF Kernel Shape:", rbf_k.shape)
```

```python
from collections import Counter
from itertools import combinations
from sklearn.svm import SVC
import numpy as np
class OvOSVM:
    def __init__(self, kernel="linear", C=1.0):
        self.kernel = kernel
        self.C = C
        self.models = []
        self.label_pairs = []

    def fit(self, X, y):
        classes = np.unique(y)
        self.models = []
        self.label_pairs = []

        for (class1, class2) in combinations(classes, 2):
            indices = (y == class1) | (y == class2)
            X_pair = X[indices]
            y_pair = y[indices]
            y_pair = np.where(y_pair == class1, 1, -1)

            model = SVC(kernel=self.kernel, C=self.C)
            model.fit(X_pair, y_pair)

            self.models.append(model)
            self.label_pairs.append((class1, class2))

    def predict(self, X):
        votes = []
        for model, (class1, class2) in zip(self.models, self.label_pairs):
            prediction = model.predict(X)
            votes.append(np.where(prediction == 1, class1, class2))

        votes = np.array(votes).T
        final_predictions = [Counter(v).most_common(1)[0][0] for v in votes]
        return np.array(final_predictions)
```

```python
def my_accuracy_score(y_true, y_pred):
    if len(y_true) != len(y_pred):
        raise ValueError("y_true vÃ  y_pred pháº£i cÃ³ cÃ¹ng sá»‘ lÆ°á»£ng pháº§n tá»­.")

    correct_predictions = sum(y_true == y_pred)

    accuracy = correct_predictions / len(y_true)
    return accuracy
```

```python
import numpy as np
from itertools import combinations
from collections import Counter


def compute_kernel(X1, X2, kernel="linear", gamma=0.01):
    if isinstance(X1, pd.DataFrame) or isinstance(X1, pd.Series):
        X1 = X1.values
    if isinstance(X2, pd.DataFrame) or isinstance(X2, pd.Series):
        X2 = X2.values

    if kernel == "linear":
        return np.dot(X1,X2.T)
    elif kernel == "rbf":
        sq_dists = np.sum(X1**2, axis=1).reshape(-1, 1) + np.sum(X2**2, axis=1) - 2 * np.dot(X1, X2.T)
        return np.exp(-gamma * sq_dists)
    else:
        raise ValueError("Kernel khÃ´ng há»£p lá»‡!")

def smo(X, y, kernel="linear", C=1.0, tol=1e-4, max_iter=10000):
    n_samples, n_features = X.shape
    K = compute_kernel(X, X, kernel)
    alpha = np.zeros(n_samples)
    b = 0

    for _ in range(max_iter):
        alpha_prev = np.copy(alpha)
        for i in range(n_samples):
            f_i = np.sum(alpha * y * K[:, i]) + b
            E_i = f_i - y[i]

            if (y[i] * E_i < -tol and alpha[i] < C) or (y[i] * E_i > tol and alpha[i] > 0):
                j = np.random.choice([x for x in range(n_samples) if x != i])
                f_j = np.sum(alpha * y * K[:, j]) + b
                E_j = f_j - y[j]

                alpha_i_old, alpha_j_old = alpha[i], alpha[j]

                if y[i] != y[j]:
                    L = max(0, alpha[j] - alpha[i])
                    H = min(C, C + alpha[j] - alpha[i])
                else:
                    L = max(0, alpha[i] + alpha[j] - C)
                    H = min(C, alpha[i] + alpha[j])

                if L == H:
                    continue
                eta = 2 * K[i, j] - K[i, i] - K[j, j]
                if eta >= 0:
                    continue
                alpha[j] -= y[j] * (E_i - E_j) / eta
                alpha[j] = np.clip(alpha[j], L, H)

                alpha[i] += y[i] * y[j] * (alpha_j_old - alpha[j])

                b1 = b - E_i - y[i] * (alpha[i] - alpha_i_old) * K[i, i] - y[j] * (alpha[j] - alpha_j_old) * K[i, j]
                b2 = b - E_j - y[i] * (alpha[i] - alpha_i_old) * K[i, j] - y[j] * (alpha[j] - alpha_j_old) * K[j, j]

                if 0 < alpha[i] < C:
                    b = b1
                elif 0 < alpha[j] < C:
                    b = b2
                else:
                    b = (b1 + b2) / 2

        diff = np.linalg.norm(alpha - alpha_prev)
        if diff < tol:
            break

    return alpha, b

# One-vs-One SVM
class myOvOSVM:
    def __init__(self, kernel="rbf", C=1.0 , gamma = 0.01):
        self.kernel = kernel
        self.C = C
        self.gamma = gamma
        self.models = []
        self.label_pairs = []

    def fit(self, X, y):
        classes = np.unique(y)
        self.models = []
        self.label_pairs = []

        for (class1, class2) in combinations(classes, 2):
            indices = (y == class1) | (y == class2)
            X_pair = X[indices]
            y_pair = y[indices]
            y_pair = np.where(y_pair == class1, 1, -1)

            alpha, b = smo(X_pair, y_pair, kernel=self.kernel, C=self.C)

            self.models.append((alpha, b, X_pair, y_pair))
            self.label_pairs.append((class1, class2))

    def predict(self, X):
      votes = []
      for model, label_pair in zip(self.models, self.label_pairs):
          alpha, b, X_train, y_train = model
          class1, class2 = label_pair

          K = compute_kernel(X, X_train, kernel=self.kernel, gamma=self.gamma)
          decision = np.sum(alpha * y_train * K, axis=1) + b
          predictions = np.sign(decision)

          for i, pred in enumerate(predictions):
              if pred == 1:
                  votes.append(class1)
              else:
                  votes.append(class2)

      votes = np.array(votes).reshape(len(self.models), -1).T
      final_predictions = [
          Counter(v).most_common(1)[0][0] for v in votes
      ]
      return np.array(final_predictions)
```

```python
my_ovo_svm = myOvOSVM(kernel="linear", C=1.0)

my_ovo_svm.fit(X_train, y_train)
```

```python
my_predictions = my_ovo_svm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy = my_accuracy_score(y_test, my_predictions)
report = classification_report(y_test, my_predictions)

print("Accuracy:", accuracy)
print("Classification Report:\n", report)
```
